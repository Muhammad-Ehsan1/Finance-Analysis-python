```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.set_style("darkgrid")
```

```python
## import Dataset
df = pd.read_csv("E:\data analytics\datasets\Finance_data.csv")
df.head()
```

```python
## Drop Unused columns
df.drop(["Mutual_Funds","Equity_Market","Debentures","Government_Bonds","PPF","Fixed_Deposits","Gold"],axis = 1 , inplace = True)
```

```python
## Replace Black values with Nan
df.replace(r'^ s*$' ,np.nan , regex=True)
```

```python
## ploting
sns.countplot(x = "gender", data= df, linewidth = 3, palette= "Set2", edgecolor = "black")
plt.show()
```

```python
sns.countplot(x = "age", data= df, linewidth = 2, palette= "Set3", edgecolor = "black")
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.countplot(x = df["gender"],hue=df["Investment_Avenues"], palette="summer", linewidth = 3, edgecolor = "white")
plt.title("Investment Avenues")

plt.subplot(1,2,2)
sns.countplot(x = df["gender"],hue=df["Stock_Marktet"], palette="hot", linewidth = 3, edgecolor = "white")
plt.title("Stock Market")
```

```python
sns.countplot(x = df["Factor"], palette="coolwarm", linewidth = 2, edgecolor = "black")
plt.show()
```

```python
plt.figure(figsize=(18,6))
sns.pointplot(x = "Purpose",y = "age",data = df ,linestyles = "--", capsize= .3 ,color="Blue" )
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.countplot(x = df["Duration"],hue=df["gender"], palette="viridis", linewidth = 2, edgecolor = "black")
plt.title("DURATION")

plt.subplot(1,2,2)
sns.countplot(x = df["Invest_Monitor"],hue=df["gender"], palette="seismic", linewidth = 2, edgecolor = "black")
plt.title("INVESTMENT MONITORING")
plt.show()
```
